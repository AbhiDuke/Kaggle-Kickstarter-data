```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

d = pd.read_csv('dspro.csv',encoding='latin1')


df = pd.DataFrame(d)
```

```python
#df.head()
```

```python
df['state']=df['state'].map({'failed':0,'successful':1})
```

```python
#df.head()
```

```python
dfx=df[df['state']<3]
#df.head()
```
